```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

df = pd.read_csv('https://raw.githubusercontent.com/Lailaamr1/CodeAlpha_StockPrices/main/AAPLL.csv')
data = df['close'].values.reshape(-1, 1)


scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

**Modelling LSTM Model**

```python
def create_dataset(dataset, time_steps=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_steps):
        a = dataset[i:(i + time_steps), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_steps, 0])
    return np.array(dataX), np.array(dataY)

time_steps = 10
X, y = create_dataset(scaled_data, time_steps)
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


# Split the data into training and testing sets
train_size = int(len(scaled_data) * 0.8)
test_size = len(scaled_data) - train_size
X_train, X_test = X[0:train_size, :], X[train_size:len(scaled_data), :]
y_train, y_test = y[0:train_size], y[train_size:len(scaled_data)]
```

```python
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(time_steps, 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

model.fit(X_train.reshape((X_train.shape[0], X_train.shape[1], 1)), y_train, epochs=80, batch_size=32)
```

```python
predictions = model.predict(X_test.reshape((X_test.shape[0], X_test.shape[1], 1)))
predictions = scaler.inverse_transform(predictions)
```

**Evaluation**

```python
# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(np.mean(np.square(predictions - df['close'].values[train_size:])))
print("Root Mean Squared Error is:",rmse)

plt.plot(df['close'].values[train_size:], label='Actual Prices')
plt.plot(predictions, label='Predicted Prices')
plt.legend()
plt.show()
```
